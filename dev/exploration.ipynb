```python
import pandas as pd
import numpy as np
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler

def prepare_fly_data(filepath, drop_se=True, drop_species=True, species_list = None):
    """
    Prepare fly trait data for analysis by cleaning and handling missing values.
    
    Parameters:
    -----------
    filepath : str
        Path to the CSV file
    drop_se : bool, default=True
        Whether to drop standard error columns
    drop_species : bool, default=True
        Whether to drop Species and comment columns
    
    Returns:
    --------
    dict containing:
        - raw_df : original DataFrame
        - clean_df : cleaned DataFrame with selected columns
        - null_summary : dictionary of null statistics
    """
    # Read data
    df = pd.read_csv(filepath)

    # Remove completely empty columns/rows (matches your new cleaning chunk)
    df = df.dropna(axis=1, how='all')
    df = df.dropna(axis=0, how='all')

    if species_list is not None and "Species" in df.columns:
        # keep order of species_list, drop missing gracefully
        found = df["Species"].isin(species_list)
        df_clean = df.loc[found].copy()
        # (optional) preserve species as index if you're subsetting by it
        df_clean = df_clean.set_index("Species")
    else:
        df_clean = df.copy()
    
    
    # Fix insemination_rxn column
   

    # Create clean dataset
    drop_cols = []
    if drop_se:
        drop_cols.extend([col for col in df.columns if col.endswith('_se')])

    df_clean = df_clean.drop(columns=drop_cols)

    if drop_species:
        df_clean = df_clean.reset_index(drop=True)
    
    # Analyze nulls
    null_cols = df_clean.isnull().sum()
    null_rows = df_clean.isnull().sum(axis=1)
    
    null_summary = {
        'total_nulls': df_clean.isnull().sum().sum(),
        'total_cells': df_clean.size,
        'null_proportion': df_clean.isnull().sum().sum() / df_clean.size,
        'cols_with_nulls': null_cols[null_cols > 0].to_dict(),
        'rows_with_nulls': (null_rows > 0).sum(),
        'row_null_dist': null_rows.describe().to_dict()
    }
    
    return {
        'raw_df': df,
        'clean_df': df_clean,
        'null_summary': null_summary
    }

def print_null_summary(null_summary):
    """Print formatted null analysis results."""
    print(f"Overall null proportion: {null_summary['null_proportion']:.2%}")
    print(f"\nRows containing nulls: {null_summary['rows_with_nulls']}")
    
    print("\nColumns with nulls (showing proportion):")
    for col, count in sorted(null_summary['cols_with_nulls'].items(), 
                           key=lambda x: x[1], reverse=True):
        print(f"{col}: {count/149:.2%}")  # 149 is total rows from metadata
    
    print("\nNull counts per row distribution:")
    for stat, value in null_summary['row_null_dist'].items():
        print(f"{stat}: {value:.2f}")

def create_analysis_ready_data(df_clean: pd.DataFrame, impute_method='knn', scale=True):
    """
    Prepare data for dimensionality reduction by handling missing values.
    
    Parameters:
    -----------
    df_clean : pandas.DataFrame
        Cleaned DataFrame
    impute_method : str, default='knn'
        Method for handling missing values ('knn' or 'drop')
    scale : bool, default=True
        Whether to standardize features
    
    Returns:
    --------
    tuple:
        - X : numpy array of prepared features
        - feature_names : list of feature names
    """
     # 1) Select numeric columns only
    numeric_df = df_clean.select_dtypes(include=[np.number]).copy()
    feature_names = numeric_df.columns.tolist()

    if numeric_df.shape[1] == 0:   # no numeric columns
        return np.empty((numeric_df.shape[0], 0)), []

    X = numeric_df.values.astype(float, copy=False)

    # 2) Handle missing values
    impute_method = impute_method.lower()
    if impute_method == "drop":
        mask = ~np.isnan(X).any(axis=1)
        X = X[mask]

    elif impute_method == "mean":
        col_means = np.nanmean(X, axis=0)
        col_means = np.where(np.isnan(col_means), 0.0, col_means)  # replace all-NaN cols
        inds = np.where(np.isnan(X))
        X[inds] = np.take(col_means, inds[1])

    elif impute_method == "knn":
        try:
            from sklearn.impute import KNNImputer
            imputer = KNNImputer(n_neighbors=5)
            X = imputer.fit_transform(X)
        except ModuleNotFoundError:
            col_means = np.nanmean(X, axis=0)
            col_means = np.where(np.isnan(col_means), 0.0, col_means)
            inds = np.where(np.isnan(X))
            X[inds] = np.take(col_means, inds[1])
    else:
        raise ValueError("impute_method must be one of: 'knn', 'mean', 'drop'")

    # 3) Scale
    if scale:
        try:
            from sklearn.preprocessing import StandardScaler
            X = StandardScaler().fit_transform(X)
        except ModuleNotFoundError:
            mean = np.mean(X, axis=0)
            std = np.std(X, axis=0, ddof=0)
            std = np.where(std == 0.0, 1.0, std)
            X = (X - mean) / std

    return X, feature_names

# Example usage:
"""
# Prepare and analyze data
results = prepare_fly_data('Fly_Traits.0.1_Jan8_2024.csv')
print_null_summary(results['null_summary'])

# Create analysis-ready data
X, features = create_analysis_ready_data(results['clean_df'])
"""
```

```python
import pandas as pd
import numpy as np
from Bio import Phylo
from io import StringIO
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial.distance import squareform, pdist
import umap
from sklearn.preprocessing import MinMaxScaler
import colorcet as cc  # For better colormaps

def load_and_process_tree(tree_file):
    """
    Load and process a Newick format tree file.
    Returns the tree and a dictionary of pairwise distances.
    """
    # Read the tree
    tree = Phylo.read(tree_file, 'newick')
    
    # Get all terminal nodes (species)
    species = [terminal.name for terminal in tree.get_terminals()]
    
    # Calculate pairwise distances
    distances = {}
    for i, sp1 in enumerate(species):
        distances[sp1] = {}
        for sp2 in species:
            if sp1 != sp2:
                dist = tree.distance(sp1, sp2)
                distances[sp1][sp2] = dist
    
    return tree, distances, species

def plot_phylogenetic_tree(tree, figsize=(12, 8)):
    """
    Create a visualization of the phylogenetic tree.
    """
    fig, ax = plt.subplots(figsize=figsize)
    Phylo.draw(tree, axes=ax, show_confidence=False)
    plt.title("Phylogenetic Tree")
    return fig

def create_distance_matrix(distances, species):
    """
    Convert distance dictionary to a matrix format.
    """
    n = len(species)
    dist_matrix = np.zeros((n, n))
    
    for i, sp1 in enumerate(species):
        for j, sp2 in enumerate(species):
            if sp1 != sp2:
                dist_matrix[i, j] = distances[sp1][sp2]
    
    return dist_matrix

def calculate_reference_species(dist_matrix, species):
    """
    Calculate a reference species (most central) based on average distances.
    Returns the species name and its index.
    """
    avg_distances = dist_matrix.mean(axis=1)
    ref_idx = np.argmin(avg_distances)
    return species[ref_idx], ref_idx

def get_phylogenetic_colors(dist_matrix, ref_idx, cmap=cc.cm.rainbow):
    """
    Generate colors based on phylogenetic distances from a reference species.
    """
    # Get distances from reference species
    distances = dist_matrix[ref_idx]
    
    # Normalize distances to [0, 1]
    scaler = MinMaxScaler()
    normalized_distances = scaler.fit_transform(distances.reshape(-1, 1)).ravel()
    
    # Convert to colors
    colors = cmap(normalized_distances)
    return colors, normalized_distances

def process_trait_data(df, species_list):
    """
    Process trait data to align with tree data and prepare for UMAP.
    """
    # Ensure species order matches tree data
    df_aligned = df.set_index('Species').loc[species_list]
    
    # Drop non-numeric columns and handle missing values
    numeric_cols = df_aligned.select_dtypes(include=[np.number]).columns
    X = df_aligned[numeric_cols]
    
    # Basic imputation (you might want to use more sophisticated methods)
    X = X.fillna(X.mean())
    
    return X

def create_umap_embedding(X, n_neighbors=15, min_dist=0.1, random_state=42):
    """
    Create UMAP embedding of trait data.
    """
    reducer = umap.UMAP(n_neighbors=n_neighbors, 
                       min_dist=min_dist, 
                       random_state=random_state)
    embedding = reducer.fit_transform(X)
    return embedding

def plot_phylo_umap(embedding, colors, species, ref_species, figsize=(10, 10)):
    """
    Create UMAP plot colored by phylogenetic distance.
    """
    fig, ax = plt.subplots(figsize=figsize)
    
    scatter = ax.scatter(embedding[:, 0], embedding[:, 1], 
                        c=colors, cmap=cc.cm.rainbow)
    
    # Add colorbar
    plt.colorbar(scatter, label='Phylogenetic Distance from ' + ref_species)
    
    # Add labels for some points (you might want to adjust this)
    for i, (x, y) in enumerate(embedding):
        if i % 5 == 0:  # Label every 5th point to avoid overcrowding
            ax.annotate(species[i], (x, y), xytext=(5, 5), 
                       textcoords='offset points', fontsize=8)
    
    plt.title('UMAP of Trait Data\nColored by Phylogenetic Distance')
    ax.set_xlabel('UMAP 1')
    ax.set_ylabel('UMAP 2')
    
    return fig

# Example usage:
"""
# Load and process tree
tree, distances, species = load_and_process_tree('your_tree.tre')

# Visualize tree
fig_tree = plot_phylogenetic_tree(tree)
plt.show()

# Create distance matrix
dist_matrix = create_distance_matrix(distances, species)

# Get reference species and colors
ref_species, ref_idx = calculate_reference_species(dist_matrix, species)
colors, _ = get_phylogenetic_colors(dist_matrix, ref_idx)

# Process trait data
trait_data = pd.read_csv('your_trait_data.csv')
X = process_trait_data(trait_data, species)

# Create UMAP embedding
embedding = create_umap_embedding(X)

# Plot UMAP with phylogenetic colors
fig_umap = plot_phylo_umap(embedding, colors, species, ref_species)
plt.show()
"""
```

```python
import pandas as pd
import numpy as np
from Bio import Phylo
import matplotlib.pyplot as plt
import seaborn as sns
import umap
from scipy.cluster.hierarchy import cut_tree, linkage
from matplotlib.collections import LineCollection
from scipy.spatial.distance import pdist

def plot_umap_with_phylo_distances(embedding, dist_matrix, species, 
                                 distance_threshold=0.5, alpha_scale=0.5,
                                 figsize=(12, 12)):
    """
    Plot UMAP with lines showing phylogenetic distances.
    
    Parameters:
    -----------
    embedding : array-like
        UMAP embedding coordinates
    dist_matrix : array-like
        Phylogenetic distance matrix
    species : list
        List of species names
    distance_threshold : float
        Maximum phylogenetic distance to show connections (as proportion of max distance)
    alpha_scale : float
        Scaling factor for line transparency
    """
    fig, ax = plt.subplots(figsize=figsize)
    
    # Normalize distances for alpha calculation
    max_dist = np.max(dist_matrix)
    threshold = max_dist * distance_threshold
    
    # Create lines and colors for phylogenetic distances
    lines = []
    colors = []
    for i in range(len(embedding)):
        for j in range(i+1, len(embedding)):
            dist = dist_matrix[i, j]
            if dist <= threshold:
                lines.append([embedding[i], embedding[j]])
                # Convert distance to alpha value (closer = more opaque)
                alpha = .5 - (dist / max_dist) * .5
                colors.append((0.7, 0.7, 0.7, alpha))  # Gray with calculated alpha
    
    # Plot distance lines
    if lines:
        lc = LineCollection(lines, colors=colors, zorder=1)
        ax.add_collection(lc)
    
    # Plot points
    scatter = ax.scatter(embedding[:, 0], embedding[:, 1], 
                        c='red', s=100, zorder=2)
    
    # Add labels
    for i, (x, y) in enumerate(embedding):
        # Adjust label density as needed
        if i % 3 == 0:  # Label every 3rd point
            ax.annotate(species[i], (x, y), xytext=(5, 5),
                       textcoords='offset points', fontsize=8)
    
    # Set plot limits based on embedding
    padding = 0.1
    x_min, x_max = embedding[:, 0].min(), embedding[:, 0].max()
    y_min, y_max = embedding[:, 1].min(), embedding[:, 1].max()
    x_range = x_max - x_min
    y_range = y_max - y_min
    ax.set_xlim(x_min - padding * x_range, x_max + padding * x_range)
    ax.set_ylim(y_min - padding * y_range, y_max + padding * y_range)
    
    plt.title('UMAP Embedding with Phylogenetic Distances')
    ax.set_xlabel('UMAP 1')
    ax.set_ylabel('UMAP 2')
    
    return fig

def plot_distance_comparison(embedding, dist_matrix, title="Distance Comparison"):
    """
    Create a scatter plot comparing UMAP distances to phylogenetic distances.
    """
    # Calculate pairwise distances in UMAP space
    umap_dists = pdist(embedding)
    phylo_dists = pdist(dist_matrix)
    
    # Create comparison plot
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.scatter(phylo_dists, umap_dists, alpha=0.1)
    ax.set_xlabel('Phylogenetic Distance')
    ax.set_ylabel('UMAP Distance')
    ax.set_title(title)
    
    # Add correlation coefficient
    corr = np.corrcoef(phylo_dists, umap_dists)[0, 1]
    ax.text(0.05, 0.95, f'Correlation: {corr:.3f}',
            transform=ax.transAxes, fontsize=10)
    
    return fig

# Example usage:
"""
# Create UMAP embedding
embedding = create_umap_embedding(X)

# Plot with phylogenetic distances
fig1 = plot_umap_with_phylo_distances(embedding, dist_matrix, species)

# Plot distance comparison
fig2 = plot_distance_comparison(embedding, dist_matrix)

plt.show()
"""
```

```python
import pandas as pd
import numpy as np
from Bio import Phylo
import matplotlib.pyplot as plt
import seaborn as sns
import umap
from matplotlib.colors import hsv_to_rgb
import colorsys

def calculate_cumulative_distances(ordered_species, dist_matrix, species_list):
    """
    Calculate cumulative distances between adjacent species in the ordered list.
    
    Parameters:
    -----------
    ordered_species : list
        Species names in phylogenetic order
    dist_matrix : array-like
        Distance matrix
    species_list : list
        Complete list of species (for index lookup)
    
    Returns:
    --------
    cumulative_distances : array
        Cumulative distances from first species
    """
    # Get indices for the ordered species
    idx_map = {sp: i for i, sp in enumerate(species_list)}
    ordered_indices = [idx_map[sp] for sp in ordered_species]
    
    # Calculate distances between adjacent species
    adjacent_distances = []
    for i in range(len(ordered_indices)-1):
        idx1, idx2 = ordered_indices[i], ordered_indices[i+1]
        adjacent_distances.append(dist_matrix[idx1, idx2])
    
    # Calculate cumulative distances
    cum_distances = np.cumsum([0] + adjacent_distances)
    
    return cum_distances

def create_hue_mapping(ordered_species, dist_matrix, species_list, hue_start=0.0, hue_end=0.8):
    """
    Create a mapping of species to hue values based on phylogenetic ordering and distances.
    
    Parameters:
    -----------
    ordered_species : list
        Species names in phylogenetic order
    dist_matrix : array-like
        Distance matrix
    species_list : list
        Complete list of species (for lookup)
    hue_start : float
        Starting point for hue range (0-1)
    hue_end : float
        Ending point for hue range (0-1)
    
    Returns:
    --------
    dict : {species_name: hue_value}
    """
    cum_distances = calculate_cumulative_distances(ordered_species, dist_matrix, species_list)
    
    # Normalize to [0, 1] first
    normalized_distances = cum_distances / cum_distances[-1]
    
    # Scale to desired hue range
    hue_range = hue_end - hue_start
    scaled_hues = (normalized_distances * hue_range) + hue_start
    
    # Create mapping
    hue_mapping = {sp: hue for sp, hue in zip(ordered_species, scaled_hues)}
    
    return hue_mapping

def plot_umap_with_phylo_hues(embedding, ordered_species, dist_matrix, species_list, 
                             hue_start=0.0, hue_end=0.8, figsize=(12, 12)):
    """
    Plot UMAP with colors based on phylogenetic ordering.
    
    Parameters:
    -----------
    embedding : array-like
        UMAP embedding coordinates
    ordered_species : list
        Species names in phylogenetic order
    dist_matrix : array-like
        Distance matrix
    species_list : list
        Complete list of species (for lookup)
    hue_start : float
        Starting point for hue range (0-1)
    hue_end : float
        Ending point for hue range (0-1)
    """
    # Create hue mapping
    hue_mapping = create_hue_mapping(ordered_species, dist_matrix, species_list, 
                                   hue_start=hue_start, hue_end=hue_end)
    
    # Create figure
    fig, ax = plt.subplots(figsize=figsize)
    
    # Create colors for each point
    colors = []
    for sp in species_list:
        hue = hue_mapping[sp]
        # Convert HSV to RGB (using maximum saturation and value)
        rgb = colorsys.hsv_to_rgb(hue, 0.8, 0.9)
        colors.append(rgb)
    
    # Plot points
    scatter = ax.scatter(embedding[:, 0], embedding[:, 1], 
                        c=colors, s=100)
    
    # Add labels
    for i, (x, y) in enumerate(embedding):
        if i % 3 == 0:  # Label every 3rd point
            ax.annotate(species_list[i], (x, y), xytext=(5, 5),
                       textcoords='offset points', fontsize=8)
    
    # Set plot limits with padding
    padding = 0.1
    x_min, x_max = embedding[:, 0].min(), embedding[:, 0].max()
    y_min, y_max = embedding[:, 1].min(), embedding[:, 1].max()
    x_range = x_max - x_min
    y_range = y_max - y_min
    ax.set_xlim(x_min - padding * x_range, x_max + padding * x_range)
    ax.set_ylim(y_min - padding * y_range, y_max + padding * y_range)
    
    plt.title('UMAP Embedding\nColors Based on Phylogenetic Ordering')
    ax.set_xlabel('UMAP 1')
    ax.set_ylabel('UMAP 2')
    
    # Add a custom colorbar showing the sequence with correct hue range
    ax_cbar = fig.add_axes([0.95, 0.1, 0.03, 0.8])
    
    # Create colorbar gradient
    gradient = np.linspace(0, 1, 256)
    colorbar_colors = np.zeros((256, 3))
    for i, val in enumerate(gradient):
        hue = val * (hue_end - hue_start) + hue_start
        colorbar_colors[i] = colorsys.hsv_to_rgb(hue, 0.8, 0.9)
    
    # Reshape and display colorbar
    gradient_array = colorbar_colors.reshape(256, 1, 3)
    ax_cbar.imshow(gradient_array, aspect='auto')
    ax_cbar.set_xticks([])
    
    # Add some species names along the colorbar
    n_labels = 5
    idx = np.linspace(0, len(ordered_species)-1, n_labels).astype(int)
    positions = np.linspace(0, 256, n_labels).astype(int)
    ax_cbar.set_yticks(positions)
    ax_cbar.set_yticklabels([ordered_species[i] for i in idx], fontsize=8)
    
    return fig

def plot_ordering_distances(ordered_species, dist_matrix, species_list, figsize=(10, 5)):
    """
    Plot distances between adjacent species in the ordering.
    """
    idx_map = {sp: i for i, sp in enumerate(species_list)}
    ordered_indices = [idx_map[sp] for sp in ordered_species]
    
    # Calculate distances between adjacent species
    distances = []
    for i in range(len(ordered_indices)-1):
        idx1, idx2 = ordered_indices[i], ordered_indices[i+1]
        distances.append(dist_matrix[idx1, idx2])
    
    # Plot
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot(distances, '-o')
    ax.set_title('Distances Between Adjacent Species in Phylogenetic Ordering')
    ax.set_xlabel('Position in Sequence')
    ax.set_ylabel('Distance to Next Species')
    
    return fig
```
